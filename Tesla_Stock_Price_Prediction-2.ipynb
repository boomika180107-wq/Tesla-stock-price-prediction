## ***5. Hypothesis Testing***

### Based on your chart experiments, define three hypothetical statements from the dataset. In the next three questions, perform hypothesis testing to obtain final conclusion about the statements through your code and statistical testing.

Answer Here.

### Hypothetical Statement - 1

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

**RESEARCH HYPOTHESIS**: There is a significant relationship between Tesla's Trading volume and closing price.

**NULL HYPOTHESIS**: There is no significant relationship between Tesla's Trading Volume and Closing price.

**ALTERNATE HYPOTHESIS**: There is a significant relationship between Tesla's Trading Volume and Closing Price.

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value
from scipy.stats import pearsonr
corr_coef, p_value = pearsonr(df1['Volume'], df1['Close'])
print("Correlation Coefficient:", corr_coef)
print("P-value:", p_value)
alpha = 0.05
if p_value < alpha:
  print("Reject the Null Hypothesis (H0)")
  print("There is a significant relationship between trading volume and closing price")
else:
  print("Fail to reject the Null Hypothesis (H0)")
  print("There is no significant relationship between trading volume and closing price")
````

##### Which statistical test have you done to obtain P-Value?

To obtain the p-value, a pearson Correlation Test was performed between Tesla's Trading Volume and Closing price. This test determines whether a statistically significant linear relationship exists between the two variables.

##### Why did you choose the specific statistical test?

I chose the Pearson Correlation Test because both trading volume and closing price are continous numerical variables. The test is suitable for measuring the strength and significance of the linear relationship between these two variables and helps determine whether the observed correlation is statstically significant.

### Hypothetical Statement - 2

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

RESEARCH HYPOTHESIS: The average Tesla Closing Price before and after the median date of the datset is significant different.

NULL HYPOTHESIS: There is no significant in the average Tesla Closing Price before and after the median date of the dataset.

ALTERNATE HYPOTHESIS: There is a significant difference in the average Tesla Closing Price before and after the median date of the dataset.

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value
from scipy.stats import ttest_ind
mid_point = len(df1) // 2
first_half = df1['Close'][:mid_point]
second_half = df1['Close'][mid_point:]
t_stat, p_value = ttest_ind(first_half, second_half)
print("T_Statstics:", t_stat)
print("P-value:", p_value)
alpha = 0.05
if p_value < alpha:
  print("Reject the Null Hypothesis(H0)")
  print("There is a significant difference in average price between the two periods")
else:
  print("Fail to reject the Null hypothesis(H0)")
  print("There is no significant difference in average closing price between the two periods.")
````

##### Which statistical test have you done to obtain P-Value?

To obtain the p-value, an independent two-samplet-test was performed on the closing prices from the first half and second half of the dataset. This test helps determine whether there is a statistically significant difference between the mean closing prices of the two periods.

##### Why did you choose the specific statistical test?

I chose the Independent Two Sample t-test because it compares the means of two independant groups(first half of the dataset and second half of the dataset) to determine whether the difference in average closing prices in statistically significant.

````python
# Perform Statistical Test to obtain P-Value
from scipy.stats import ttest_1samp
daily_returns = df1['Close'].pct_change().dropna()
t_stat, p_value = ttest_1samp(daily_returns, 0)
print("T-statistic", t_stat)
print("P-value", p_value)
alpha = 0.05
if p_value < alpha:
  print("Reject the Null Hypothesis (H0)")
  print("The mean daily return is significantly different from zero")
else:
   print("Fail to reject the Null hypothesis(H0)")
   print("The mean daily return is not significantly different from zero")

````

To obtain the p-value, a one sample t-test was performed on Tesla's daily returns. This test evaluates whether the avergae daily return is statistically different from zero.

### 3. Explain the model which you have used and the feature importance using any model explainability tool?

### Explanation of the GRU Model:

The **Gated Recurrent Unit (GRU)** is a type of recurrent neural network (RNN) that is particularly effective for sequential data like time series. It was introduced to address the vanishing gradient problem that can occur in standard RNNs, which makes it difficult for them to learn long-term dependencies.

**Key Components of a GRU:**

Unlike an LSTM which has three gates (input, forget, and output), a GRU has two gates:

1.  **Update Gate (z_t)**: This gate decides how much of the past information (from the previous hidden state) needs to be passed along to the future. It acts like a combination of the input and forget gates of an LSTM. A high value for the update gate means retaining more of the past information.
2.  **Reset Gate (r_t)**: This gate determines how much of the previous hidden state to forget. A low value for the reset gate means forgetting more of the past. It effectively allows the model to drop information that is irrelevant to the future prediction.

**How it Works (Simplified):**

*   At each time step, the GRU considers the current input and the previous hidden state.
*   The **reset gate** decides which parts of the previous hidden state are relevant to the current input.
*   The **update gate** then controls how much of the new information (based on the current input and the 'reset' previous hidden state) and how much of the old information (previous hidden state) should be combined to form the new hidden state.
*   This mechanism allows GRUs to selectively remember and forget information over long sequences, making them powerful for tasks like stock price prediction where past trends and events can influence future prices.

**Advantages:**

*   **Handles Long-Term Dependencies**: Effectively learns from data points far apart in a sequence.
*   **Simpler Architecture**: Fewer gates than LSTMs, leading to less complexity and potentially faster training, while often achieving comparable performance.
*   **Reduced Vanishing Gradient Problem**: The gating mechanism helps mitigate the vanishing gradient issue.

### Feature Importance:

For deep learning models like GRU, directly obtaining explicit 'feature importance' scores in the same way as, say, tree-based models (e.g., Random Forest, XGBoost) is not straightforward. The model learns complex, non-linear relationships across the input sequence.

However, we can infer the importance of the input features (`Open`, `High`, `Low`, `Volume`) based on:

1.  **Domain Knowledge**: In financial markets, these four features are fundamentally important. The `Open`, `High`, and `Low` prices define the trading range and immediate sentiment of the day, while `Volume` indicates the level of trading activity and conviction behind price movements.
2.  **Correlation Analysis (Chart - 3)**: Our earlier correlation heatmap showed a very high positive correlation between `Open`, `High`, `Low`, and `Close` prices. This statistically confirms their strong relationship and relevance to the target variable.
3.  **Model Input Structure**: The GRU model is designed to process these features over a sequence. While not explicit scores, the model inherently leverages information from all these inputs to make predictions. Features that contribute more to reducing the loss during training are implicitly more 'important' to the model's predictions.

**Using Model Explainability Tools (Conceptual):**

To explicitly quantify feature importance for a GRU model, more advanced techniques such as **Permutation Importance**, **SHAP (SHapley Additive exPlanations)**, or **LIME (Local Interpretable Model-agnostic Explanations)** would be required. These tools work by analyzing how perturbing input features affects the model's output or by approximating the model's behavior with simpler, interpretable models. Implementing these tools is beyond the current scope but would provide a more quantitative understanding of feature importance.
